```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
## 1. Fig 4b, Fig 4c --- ---
```

```python
def get_bar_plot(transcript):

    dir = '/home/han/proj_het_AC/rerun_analysis/0_Manuscript_codes_submission_20250723/Figure4/Data/' 
    data = pd.read_csv(dir+'3UTR_population_changes.txt', sep='\t')
    flt = data[data['transcript'] == transcript]


    list_pct = [(flt.cluster0_Nolig.values[0])*2,(flt.cluster1_Nolig.values[0])*2]
    list_lable = ['cluster_0_30C','cluster_1_30C']
    list_group = ['cluster0','cluster1']


    df  = pd.DataFrame({'temp' : list_group, 
                                'cluster' : list_lable,
                                'percent' :list_pct}, 
                                columns=['temp','cluster','percent'])

    df['percent'] = df['percent']
    df  = pd.DataFrame(df)
    df['temp'] = df['temp'].astype(str)

    ### plot
    fig = plt.figure(figsize = (3, 4))
    ax = sns.barplot(x=df.temp,y=df.percent,data=df)

    for c in ax.containers:
                
        # Optional: if the segment is small or 0, customize the labels
        labels = [v.get_height() if v.get_height() > 0 else '' for v in c]   

        # remove the labels parameter if it's not needed for customized labels
        #ax.bar_label(c, labels=labels, label_type='center')
        ax.bar_label(c, fmt='%0.1f%%', label_type='center', color='black', fontsize = 14)  # fmt='%.2f%%', percentage

    plt.xlabel(" ")
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.ylabel("Proportion %",fontsize=14)
    plt.title(transcript+"_\n 30C",fontsize=16)
    dir_out = '/home/han/proj_het_AC/rerun_analysis/0_Manuscript_codes_submission_20250723/Figure4/Data/'
    plt.savefig(dir_out+'Fig4b_4c_validate_barplot_30C_'+transcript+'.pdf',bbox_inches='tight', dpi=300)
    plt.show()
  


    ### 37C

    list_pct = [(flt.cluster0_Lig.values[0])*2,(flt.cluster1_Lig.values[0])*2]
    list_lable = ['cluster_0_37C','cluster_1_37C']
    list_group = ['cluster0','cluster1']


    df  = pd.DataFrame({'temp' : list_group, 
                                'cluster' : list_lable,
                                'percent' :list_pct}, 
                                columns=['temp','cluster','percent'])

    df['percent'] = df['percent']
    df  = pd.DataFrame(df)
    df['temp'] = df['temp'].astype(str)

    ### plot
    fig = plt.figure(figsize = (3, 4))
    ax = sns.barplot(x=df.temp,y=df.percent,data=df)

    for c in ax.containers:
                
        # Optional: if the segment is small or 0, customize the labels
        labels = [v.get_height() if v.get_height() > 0 else '' for v in c]   

        # remove the labels parameter if it's not needed for customized labels
        #ax.bar_label(c, labels=labels, label_type='center')
        ax.bar_label(c, fmt='%0.1f%%', label_type='center', color='black', fontsize = 14)  # fmt='%.2f%%', percentage

    plt.xlabel(" ")
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.ylabel("Proportion %",fontsize=14)
    plt.title(transcript+"_\n 37C",fontsize=16)
    dir_out = '/home/han/proj_het_AC/rerun_analysis/0_Manuscript_codes_submission_20250723/Figure4/Data/'
    plt.savefig(dir_out+'Fig4b_4c_validate_barplot_37C_'+transcript+'.pdf',bbox_inches='tight', dpi=300) 

```
