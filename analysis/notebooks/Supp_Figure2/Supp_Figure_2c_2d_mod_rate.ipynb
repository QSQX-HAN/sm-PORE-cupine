### Summary

```python
df_results = pd.read_csv("/home/han/proj_het_AC/rerun_analysis/0_Manuscript_codes_submission_20250723/Supp_Fig2/Data/Summary_ModRate_diff_align_method.csv", sep=',')
sort_order = ["100mM_1AI", "35mM_NAIN3", "50mM_NAIN3", "50mM_kethoxal", "Unmodified"] 
```

```python
chart1 = alt.Chart(df_results).mark_boxplot().encode(
    x=alt.X("method", axis=alt.Axis(title=None)),
    y=alt.Y("modrate", axis=alt.Axis(title="Mod Rate (%)")),
    color=alt.Color("method", legend=alt.Legend(title="Method")),
    column=alt.Column("batch", title="Modification rate per transcript", sort=sort_order) #Probing Chemistry and Concentration
).transform_filter(
    (datum.condition == "WT")
).properties(
    width=100,
    height=200
)

chart1.configure_axis(
    labelFontSize=12,
    titleFontSize=14
).configure_legend(
    labelFontSize=12,
    titleFontSize=14
).configure_header(
    labelFontSize=12,
    titleFontSize=14
)
chart1.encoding.y.title = 'Mod Rate (%)'
display(chart1)

save(chart1, "/home/han/proj_het_AC/rerun_analysis/0_Manuscript_codes_submission_20250723/Supp_Fig2/Data/Supp_Figure_2c_mod_rate_per_transcript_diff_methods.pdf")  
```

```python
chart2 = alt.Chart(df_results).mark_boxplot().encode(
    x=alt.X("method", axis=alt.Axis(title=None)),
    y=alt.Y("SNR", axis=alt.Axis(title="Fold Change")),
    color=alt.Color("method", legend=alt.Legend(title="Method")),
    column=alt.Column("batch", title="Fold Enrichment over Background", sort=sort_order)
).transform_filter(
    (datum.condition == "WT") & (datum.batch != "Unmodified")
).properties(
    width=100,
    height=200
)

chart2.configure_axis(
    labelFontSize=12,
    titleFontSize=14
).configure_legend(
    labelFontSize=12,
    titleFontSize=14
).configure_header(
    labelFontSize=12,
    titleFontSize=14
)

chart2.encoding.y.title = 'Fold Change'
display(chart2)

save(chart2, "/home/han/proj_het_AC/rerun_analysis/0_Manuscript_codes_submission_20250723/Supp_Fig2/Data/Supp_Figure_2d_fold_change_mod_diff_methods.pdf")  
```
