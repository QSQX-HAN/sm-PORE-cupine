```python
def plot_individual(ribo):

    dir = "/home/han/proj_het_AC/rerun_analysis/0_Manuscript_codes_submission_20250723/Figure1/Data/"
    mod = pd.read_csv(dir+"summary_mod_rate_profile_minimap2_vs_DTW.txt", sep='\t')
    mod = mod.drop_duplicates()

    dplot = mod[mod['gene'] == ribo]
    dplot['mod_rate_minimap2'] =  dplot['mod_rate_minimap2'].astype(float)
    dplot['mod_rate_dtw'] =  dplot['mod_rate_dtw'].astype(float)
    dplot['position'] =  dplot['position'].astype(int)

    r, pval = spstats.pearsonr(dplot.mod_rate_minimap2, dplot.mod_rate_dtw)
    print(r)

    fig, ax = plt.subplots(1, 1, figsize=(12, 4), dpi=100)
    ax.bar(dplot.position, dplot.mod_rate_minimap2, label="Minimap2 (n=%s,%.2f%%)" % (list(dplot.freq_mm2)[0],(dplot.mod_rate_minimap2.mean())*100), width=1.0)
    ax.bar(dplot.position, -dplot.mod_rate_dtw, label="DTW (n=%s,%.2f%%)" % (list(dplot.freq_dtw)[0],(dplot.mod_rate_dtw.mean())*100), width=1.0)
    ax.legend()
    ax.set_yticks([0.2, 0.0, -0.2])
    ax.set_yticklabels([0.2, 0.0, 0.2])
    ax.set_xlabel("Position")
    ax.set_ylabel("Mod Rate")
    ax.set_title("%s (r=%.3f)" % (ribo, r))

    SMALL_SIZE = 8
    MEDIUM_SIZE = 14
    BIGGER_SIZE = 16

    plt.rc('font', size=MEDIUM_SIZE)          # controls default text sizes
    plt.rc('axes', titlesize=MEDIUM_SIZE)     # fontsize of the axes title
    plt.rc('axes', labelsize=MEDIUM_SIZE)     # fontsize of the x and y labels
    plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
    plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
    plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize
    plt.rc('figure', titlesize=BIGGER_SIZE)   # fontsize of the figure title

    dir_out = '/home/han/proj_het_AC/rerun_analysis/0_Manuscript_codes_submission_20250723/Figure1/Data/'
    plt.savefig(dir_out+'revision_sm-PORE-cupine_dtw_mm2_modrate_correlation_'+ribo+'.pdf',bbox_inches='tight')
```

```python
plot_individual('ADD-TRUNC')
plot_individual('DEN-WT')
plot_individual('FTL6-WT')
plot_individual('GUA1')
plot_individual('LYSC')
plot_individual('MRPS21-Father')
plot_individual('RIBD')
plot_individual('SAM1')
plot_individual('TETRA-WT')
plot_individual('TPP1-ThiM')
plot_individual('YDAO')
plot_individual('YKOK')
plot_individual('YVRC')
```
