```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
###RNA002 
RNA002_NAIN3 =  "/home/han/proj_het_AC/rerun_analysis/0_Manuscript_codes_submission_20250723/Supp_Fig11/Data/RNA002_50mMNAIN3_throughput.csv"
RNA002_DMSO = "/home/han/proj_het_AC/rerun_analysis/0_Manuscript_codes_submission_20250723/Supp_Fig11/Data/RNA002_DMSO_throughput.csv"
NAIN3_R2 = pd.read_csv(RNA002_NAIN3, sep=',')
DMSO_R2 = pd.read_csv(RNA002_DMSO, sep=',')

###RNA004
RNA004_NAIN3 =  "/home/han/proj_het_AC/rerun_analysis/0_Manuscript_codes_submission_20250723/Supp_Fig11/Data/RNA004_50mMNAIN3_throughput.csv"
RNA004_DMSO = "/home/han/proj_het_AC/rerun_analysis/0_Manuscript_codes_submission_20250723/Supp_Fig11/Data/RNA004_DMSO_throughput.csv"
NAIN3_R4 = pd.read_csv(RNA004_NAIN3, sep=',')
DMSO_R4 = pd.read_csv(RNA004_DMSO, sep=',')

###RNA002 
NAIN3_R2['Time(Hours)'] = NAIN3_R2['Experiment Time (minutes)']/60
NAIN3_R2['Total'] = NAIN3_R2['Basecalled Reads Passed'] + NAIN3_R2['Basecalled Reads Failed']
NAIN3_R2['Passed'] = NAIN3_R2['Basecalled Reads Passed']
NAIN3_R2['Failed'] = NAIN3_R2['Basecalled Reads Failed']

DMSO_R2['Time(Hours)'] = DMSO_R2['Experiment Time (minutes)']/60
DMSO_R2['Total'] = DMSO_R2['Basecalled Reads Passed'] + DMSO_R2['Basecalled Reads Failed']
DMSO_R2['Passed'] = DMSO_R2['Basecalled Reads Passed']
DMSO_R2['Failed'] = DMSO_R2['Basecalled Reads Failed']

###RNA004
NAIN3_R4['Time(Hours)'] = NAIN3_R4['Experiment Time (minutes)']/60
NAIN3_R4['Total'] = NAIN3_R4['Basecalled Reads Passed'] + NAIN3_R4['Basecalled Reads Failed']
NAIN3_R4['Passed'] = NAIN3_R4['Basecalled Reads Passed']
NAIN3_R4['Failed'] = NAIN3_R4['Basecalled Reads Failed']

DMSO_R4['Time(Hours)'] = DMSO_R4['Experiment Time (minutes)']/60
DMSO_R4['Total'] = DMSO_R4['Basecalled Reads Passed'] + DMSO_R4['Basecalled Reads Failed']
DMSO_R4['Passed'] = DMSO_R4['Basecalled Reads Passed']
DMSO_R4['Failed'] = DMSO_R4['Basecalled Reads Failed']

NAIN3_R2['group'] = 'RNA002 50mMNAIN3'
DMSO_R2['group'] = 'RNA002 DMSO'
NAIN3_R4['group'] = 'RNA004 50mMNAIN3'
DMSO_R4['group'] = 'RNA004 DMSO'
dplot = pd.concat([NAIN3_R2,DMSO_R2,NAIN3_R4,DMSO_R4],ignore_index=True)
dplot['Total'] = dplot['Total']/1000000
dplot['Passed'] = dplot['Passed']/1000000
dplot['Failed'] = dplot['Failed']/1000000
dplot['Basecalled Bases'] = dplot['Basecalled Bases']/1000000
```

```python
plt.figure(figsize=(6,5))
g=sns.lineplot(data=dplot, x='Time(Hours)', y='Basecalled Bases', hue='group', palette=['orange', 'lightgreen', 'darkred', 'darkgreen'])  ### palette=['red', 'blue', 'purple', 'pink']
sns.despine()
plt.xlim([1,40])
plt.xlabel('Time(Hours)',fontsize = 16)
plt.ylabel('Sequenced Bases(M)',fontsize = 20)
plt.title(" ",fontsize = 20)
sns.move_legend(g,"upper left",fontsize = 11)
dir_out = '/home/han/proj_het_AC/rerun_analysis/0_Manuscript_codes_submission_20250723/Supp_Fig11/Data/'
plt.savefig(dir_out+'Supp_Figure_11a_RNA004_reads_yield_across_time_Bases.pdf', bbox_inches='tight', dpi=300) 
```
