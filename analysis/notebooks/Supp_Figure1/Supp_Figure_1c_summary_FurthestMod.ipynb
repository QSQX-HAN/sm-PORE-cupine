```python
dir_out = "/home/han/proj_het_AC/rerun_analysis/0_Manuscript_codes_submission_20250723/Supp_Fig1/Data/"
df_max_length = pd.read_csv(dir_out+"summary_modified_distance.txt",sep="\t")
```

```python
sort_order = ["100mM_1AI", "35mM_NAIN3", "50mM_NAIN3", "50mM_kethoxal", "Unmodified"]

fig, ax = plt.subplots(figsize=(5, 4), dpi=100)
for no, p_batch in enumerate(sort_order):  # doing this because values are not equal
    sns.kdeplot(data=df_max_length[df_max_length.batch==p_batch], x="max_length", label=p_batch, ax=ax, color=altair_colors[no])

ax.legend(loc='upper right', bbox_to_anchor=(1.5, 1))
ax.set_xlabel("Relative Max Distance (% of Gene Length)", fontsize=14)
ax.set_ylabel("Density", fontsize=14)
ax.set_title("Distance between furthest 2 modifications")      #Distance between furthest two modifications

dir_out = "/home/han/proj_het_AC/rerun_analysis/0_Manuscript_codes_submission_20250723/Supp_Fig1/Data/"
fig.savefig(dir_out+'Supp_Figure_1c_distance_furthest_modification.pdf',bbox_inches='tight')
```
