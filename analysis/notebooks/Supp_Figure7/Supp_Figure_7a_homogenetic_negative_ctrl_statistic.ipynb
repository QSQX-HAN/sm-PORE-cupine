```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
dir_out = '/home/han/proj_het_AC/rerun_analysis/0_Manuscript_codes_submission_20250723/Supp_Fig7/Data/'
dfall = pd.read_csv(dir_out+'full_len_window_homo.txt',sep='\t')   

flt = dfall.dropna() 
dplot = flt 
quan1 = 0.8
diff1 = dplot['homogeneity'].quantile(q=quan1)
quan2 = 0.2
diff2 = dplot['homogeneity'].quantile(q=quan2)
print(diff1)
print(diff2)


### plots -------------------------

plt.figure(figsize=(6,2), dpi=100)

ax=sns.histplot(dplot.homogeneity, color='steelblue', fill=True,alpha=.3,linewidth=1)     #  label="30C (N=%s)" % (counts[1]), 

### top5%
ax.axvline(dplot['homogeneity'].quantile(q=quan1),color='red',linewidth=1.5,linestyle='dotted')
ax.text(dplot['homogeneity'].quantile(q=quan1)+0.01, y=200, s="top 20% \n homogeneity=0.998",fontsize = 8)

### bottom5%
ax.axvline(dplot['homogeneity'].quantile(q=quan2),color='green',linewidth=1.5,linestyle='dotted')
ax.text(dplot['homogeneity'].quantile(q=quan2)-0.050, y=200, s="bottom 20% \n homogeneity=0.953",fontsize = 8)

plt.xlabel("Homogeneity")
plt.legend(loc="upper left", fontsize=9)
plt.title("Homogeneity")

dir_out = '/home/han/proj_het_AC/rerun_analysis/0_Manuscript_codes_submission_20250723/Supp_Fig7/Data/'
plt.savefig(dir_out+'Supp_Figure_7a_sarscov2_homo_distribution.pdf',bbox_inches='tight', dpi=300) 
```
