```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from scipy import stats
```

```python
## 1. Supp_Figure_9a ---  
```

```python
def get_replicate_corr(replicate_1,replicate_2):

    out = '/home/han/proj_het_AC/rerun_analysis/0_Manuscript_codes_submission_20250723/Supp_Fig9/Data/'
    read_count = pd.read_csv(out+'All_samples_each_count_all.csv',sep='\t')
    read_count['FPKM_log'] = np.log2(read_count['FPKM'])

    rep_1 = read_count[read_count['group'] == replicate_1]
    rep_2 = read_count[read_count['group'] == replicate_2]
    rep_1 = rep_1[['transcript_id','FPKM_log']]
    rep_2 = rep_2[['transcript_id','FPKM_log']]
    df_merge = pd.merge(rep_1,rep_2,how='inner',on=['transcript_id','transcript_id'])
    df_merge.replace([np.inf, -np.inf], np.nan, inplace=True)
    df_merge = df_merge.dropna()

    plt.figure(figsize=(7,7))
    sns.scatterplot(data=df_merge,  x=df_merge.FPKM_log_x, y=df_merge.FPKM_log_y,  s=100, alpha=0.2)    #color="steelblue"
    sns.despine()
    
    res = stats.pearsonr(df_merge.FPKM_log_x, df_merge.FPKM_log_y)
    print(res[0])

    plt.title("r=""{:.3f}".format(res[0]),fontsize=20)

    plt.xlabel(replicate_1+'(log2FPKM)',fontsize=16)
    plt.ylabel(replicate_2+'(log2FPKM)',fontsize=16)
    plt.xticks(fontsize = 12) 
    plt.yticks(fontsize = 12) 

    dir_out = '/home/han/proj_het_AC/rerun_analysis/0_Manuscript_codes_submission_20250723/Supp_Fig9/Data/'
    plt.savefig(dir_out+'Supp_Figure_9a_decay_corr_'+replicate_1+'_vs_'+replicate_2+'.pdf',bbox_inches='tight', dpi=300) 
```

```python
get_replicate_corr('30C_0min_Rep1','30C_0min_Rep2')
get_replicate_corr('30C_20min_Rep1','30C_20min_Rep2')
get_replicate_corr('30C_40min_Rep1','30C_40min_Rep2')
get_replicate_corr('30C_60min_Rep1','30C_60min_Rep2')
get_replicate_corr('30C_120min_Rep1','30C_120min_Rep2')


get_replicate_corr('37C_0min_Rep1','37C_0min_Rep2')
get_replicate_corr('37C_20min_Rep1','37C_20min_Rep2')
get_replicate_corr('37C_40min_Rep1','37C_40min_Rep2')
get_replicate_corr('37C_60min_Rep1','37C_60min_Rep2')
get_replicate_corr('37C_120min_Rep1','37C_120min_Rep2')
```
